```python
day = ['월','화','수','목','금','토','일']
data =[]
for i in day:
    noon = df[((df['주야']=='주')|(df['주야']=='주간'))&(df['요일']==i)].count()['주야']
    night = df[((df['주야']=='야')|(df['주야']=='야간'))&(df['요일']==i)].count()['주야']
    data.append([i, night, noon])
```
